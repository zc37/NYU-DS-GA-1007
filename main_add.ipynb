```python
import pandas as pd
import numpy as np
import googlemaps
import pandas as pd
from datetime import datetime
df_main = pd.read_csv('clean_main.csv')
```

```python
df_main['commercial_normal'] =100*((df_main['commercial_units'] - df_main['commercial_units'].min())
                               /(df_main['commercial_units'].max() - df_main['commercial_units'].min()))
df_main['commercial_score'] = 0.03*df_main['commercial_normal']
df_main['commercial_score'] = df_main['commercial_score'].fillna(value=0)
```

```python
def get_taxRate(x):
    Class = str(list(x)[0])
    if Class == '1':
        return 0.20919
    if Class == '2':
        return 0.12612
    if Class == '3':
        return 0.12093
    if Class == '4':
        return 0.10514
```

```python
df_main['tax_rate'] = df_main['tax_class_at_present_code'].apply(get_taxRate)
df_main['tax_normal'] = 100-(100*((df_main['tax_rate'] - df_main['tax_rate'].min())
                               /(df_main['tax_rate'].max() - df_main['tax_rate'].min())))
df_main['tax_score'] = 0.05*df_main['tax_normal']
df_main['tax_score'] = df_main['tax_score'].fillna(value=0)
```

```python
df_main2 = df_main.dropna(subset=['sale_price','floor_area_total_building']).copy()
```

```python
df_main2['unit_price'] = df_main2['sale_price'] / df_main2['floor_area_total_building']
```

```python
#df_main3[['borough_code_definition','neighborhood','sale_price','floor_area_total_building','unit_price']]
```

```python
#df_main2['unit_price'].value_counts().sort_index()
df_main3 = df_main2[df_main2.sale_price > 10000].copy()
```

```python
df_main3['price_normal'] = 100-(100*((df_main3['unit_price'] - df_main3['unit_price'].min())
                               /(df_main3['unit_price'].max() - df_main3['unit_price'].min())))
df_main3['price_score'] = 0.2*df_main3['price_normal']
#df_main3
```

```python
df_main3['property_age']= 2018 - df_main3['year_built']
df_main3['property_quality_normal'] = 100-(100*((df_main3['property_age'] - df_main3['property_age'].min())
                               /(df_main3['property_age'].max() - df_main3['property_age'].min())))
df_main3['property_quality_score'] = 0.12*df_main3['property_quality_normal']
df_main3['property_quality_score'] = df_main3['property_quality_score'].fillna(value=0)
df_main3
```
