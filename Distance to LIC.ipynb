```python
import googlemaps
import numpy as np
import pandas as pd
from datetime import datetime

gmaps = googlemaps.Client(key='AIzaSyCH7_9xqj7OH7aBIoM7MFGFix-DXJ3O54Y')
df_main = pd.read_csv('new_main.csv')
'''此py根据主表各个房产地址信息得到各个房产到达LIC的时间（开车或者坐公共交通）
此py遇到了利用google map direction功能运行效率低，得不到结果的问题
后来此py利用distance_matrix可以一次性输入100个出发地址的特点，提高了数据处理效率'''
```

```python
l = df_adrs.address.tolist()
# add_list = l
# drive_time_list = []
date = datetime.strptime('2019, 1, 15, 8, 30','%Y, %m, %d, %H, %M')
def chunks(l, n): #此函数用于将地址分为多个 含100条信息的list（为了后面distance_matrix使用）
    """Yield successive n-sized chunks from l."""
    for i in range(0, len(l), n):
        yield l[i:i + n]

def drive_time(List): #此函数返回每个地址到LIC开车时间（时间定在工作日的8点半，因为我认为这是一个常见的上班时间）
    add_list = List
    drive_time_list = []
    l = List
    for i in chunks(l, 100):
        y = gmaps.distance_matrix(i,"Long Island City", mode="driving",departure_time=date)
        for x in y['rows']:
            try:
                drive_time_list.append(x['elements'][0]['duration_in_traffic']['value'])
            except:
                drive_time_list.append(None)

    d = dict(zip(add_list,drive_time_list))  
    return d

d = drive_time(l)
pd.DataFrame.from_dict(d,orient='index')
```

```python
df_driveTime = pd.DataFrame.from_dict(d,orient='index').dropna() #去除缺少信息的row
df_driveTime.columns = ['Drive_Time']
df_driveTime['NormalValue'] = 100-(100*((df_driveTime['Drive_Time'] - df_driveTime['Drive_Time'].min())
                               /(df_driveTime['Drive_Time'].max() - df_driveTime['Drive_Time'].min()))) #打分

df_driveTime['Drive_Time_Score'] = 0.06*df_driveTime['NormalValue']
#df_driveTime
df_driveTime.to_csv('Drive_Time.csv')
```

```python
l = df_adrs.address.tolist()
#add_list = [x.split(',')[0] for x in l]
#drive_time_list = []
date = datetime.strptime('2019, 1, 15, 8, 30','%Y, %m, %d, %H, %M')
def transit_time(List): #此函数得到各个地址到LIC坐公交的时间，下面的处理与开车类似
    add_list = List
    drive_time_list = []
    l = List
    for i in chunks(l, 100):
        y = gmaps.distance_matrix(i,"Long Island City", mode="transit",departure_time=date)
        for x in y['rows']:
            try:
                drive_time_list.append(x['elements'][0]['duration']['value'])
            except:
                drive_time_list.append(None)

    d = dict(zip(add_list,drive_time_list))  
    return d

d2 = transit_time(l)
pd.DataFrame.from_dict(d2,orient='index')
```

```python
df_transitTime = pd.DataFrame.from_dict(d2,orient='index')
df_transitTime.columns = ['Transit_Time']
df_transitTime['NormalValue'] = 100-(100*((df_transitTime['Transit_Time'] - df_transitTime['Transit_Time'].min())
                               /(df_transitTime['Transit_Time'].max() - df_transitTime['Transit_Time'].min())))

df_transitTime['Transit_Time_Score'] = 0.06*df_transitTime['NormalValue']
#df_transitTime
df_transitTime.to_csv('Transit_Time.csv')
```
