```python
import googlemaps
import pandas as pd
from datetime import datetime
gmaps = googlemaps.Client(key='AIzaSyDA1tcl8xYJGOwam-Ut5VPEn5T5lRRqAac')
df_main = pd.read_csv('NewYorkCityPropertySales.csv')
'''此py将各学校的教学质量表转化为各neighbr学校2016年教学质量表'''

def get_nghbr_edu(dbn): #此函数利用googlemap获得各学校属于哪个neighbr
    try:
        geocode_result = gmaps.geocode('%s'%(dbn))   #将地址转换为经纬度
        geoInfo = geocode_result[0]['geometry']['location']
        geocode = (geoInfo['lat'],geoInfo['lng'])
        reverse_geocode_result = gmaps.reverse_geocode(geocode,result_type='neighborhood') #利用经纬度求出neighbr
        neghbr = reverse_geocode_result[0]['address_components'][0]['long_name'].upper()
        return neghbr
    except:
        return None
    
def get_score(rate): # 此函数给教育质量打分
    if rate == 'Exceeding Target':
        return 1
    elif rate == 'Meeting Target':
        return 0.75
    elif rate == 'Approaching Target':
        return 0.5
    elif rate == 'Not Meeting Target':
        return 0.25
    else:
        return None
```

```python
df_edu = pd.read_csv('EduSQR.csv',index_col=[0])
df_edu = df_edu[['DBN','Student Achievement Rating','year']] #提取需要信息
df_edu = df_edu.loc[df_edu['year'] == 2016].dropna() #取出2016年信息并且去除缺少信息的信息条
df_edu['neighborhood'] = df_edu['DBN'].apply(get_nghbr_edu).dropna() #增加neighbr 行
df_edu['EduScore'] = df_edu['Student Achievement Rating'].apply(get_score).dropna()#打分

gb1 = df_edu['EduScore'].groupby(df_edu['neighborhood'])  #以nieghbr 进行groupby
df_final_edu = pd.DataFrame(gb1.mean()) #求每个neighbr学校质量的平均数
df_test = df_main
df_result = pd.merge(df_test,df_final_edu,how='left',on='neighborhood')    
```
