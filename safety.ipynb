```python
import pandas as pd
import numpy as np

safety_dict = {}
with open("crime.csv", 'r') as f:
    for i in f.readlines()[1:]:
        key, value = list(map(lambda x: x.replace("\n", ""), i.split(",")))
        safety_dict[key] = value
df = pd.read_excel('data.xlsx')
newDf = pd.DataFrame(df, columns=['police_precinct'])
new_col = []
for index, row in newDf.iterrows():
    try:
        key = str(int(row['police_precinct']))
        new_col.append([safety_dict[key]])
    except:
        new_col.append([""])
df['CrimeNumber'] = pd.DataFrame(new_col)
df.to_csv("data.csv")
#print(df.loc[:, ['police_precinct', ' CrimeNumber ']])
df2 = pd.read_csv('data.csv')
df3 = df2.loc[ : , ['police_precinct', 'CrimeNumber']]
print(df3)

df4 = df3.sort_index(axis = 0,ascending = True,by = 'police_precinct')
print(df4)

```
